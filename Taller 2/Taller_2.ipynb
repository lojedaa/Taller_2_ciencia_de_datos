```python
bin_cols = [
    'jacuzzi','chimenea','permite_mascotas','gimnasio','ascensor','conjunto_cerrado',
    'piscina','salon_comunal','terraza','vigilancia',
    'is_cerca_estacion_tm','is_cerca_parque', 'piso', 'closets'
]

for col in bin_cols:
    print(col, df_ini[col].dropna().unique())
```

todos se pueden transformar a bolean aunque hay unos que no hay y otros que solo hay de un tipo es decir que no valen la pena mantenerlos se hara la transformacion generla y luego se eliminarna dichas columnas


```python
columnas_booleanas = []
for columna in df_ini.columns:
    valores = set(df_ini[columna].dropna().unique())
    if valores.issubset({0, 1, 0.0, 1.0, True, False}):
        columnas_booleanas.append(columna)

columnas_numericas = [
    "codigo",
    "precio_venta",
    "precio_arriendo",
    "area",
    "habitaciones",
    "banos",
    "administracion",
    "parqueaderos",
    "estrato",
    "piso",
    "closets",
    "latitud",
    "longitud",
    "distancia_estacion_tm_m",
    "distancia_parque_m"
]

columnas_fecha = [
    "datetime",
    "last_view"
]

columnas_texto = [
    "tipo_propiedad",
    "tipo_operacion",
    "sector",
    "estado",
    "localidad",
    "barrio",
    "parque_cercano",
    "estacion_tm_cercana",
    "antiguedad",
    "descripcion",
    "direccion",
    "compañia",
    "website",
    "url"
]


for col in columnas_booleanas:
    df_ini[col] = df_ini[col].astype(bool)



for col in columnas_numericas:
    df_ini[col] = pd.to_numeric(df_ini[col], errors='coerce')



for col in columnas_fecha:
    df_ini[col] = pd.to_datetime(df_ini[col], errors='coerce')


for col in columnas_texto:
    df_ini[col] = df_ini[col].astype('object')


```

```python
columnas_eliminar = [
    "_id",
    "url",
    "website",
    "descripcion",
    "direccion",
    "timeline",
    "compañia",
    "coords_modified",
    "piso",
    "closets",
    "website",
    "codigo"
    ""
]

df_ini = df_ini.drop(columns=columnas_eliminar, errors="ignore")


df_ini = df_ini[~df_ini["precio_venta"].isna()]
```

Las siguientes variables fueron eliminadas porque no aportan información útil para predecir el precio del inmueble, presentan altos porcentajes de valores nulos o corresponden a información técnica del proceso de extracción que no describe al apartamento:
* _id y codigo: identificadores del registro, no representa ninguna característica
del inmueble y ya tenemos codigo como identificador único.
* url: enlace del anuncio; información externa irrelevante para el modelo.
* website: portal donde se publicó el anuncio, no influye en el valor real del * inmueble y solo añade ruido.
* descripcion: texto libre no estructurado, continee informacion que ya hay en otras columnas.
* direccion: texto difícil de estandarizar y con más del 70% de valores nulos, además se reemplaza por latitud y longitud.
* compañía: inmobiliaria responsable del anuncio, no describe características del inmueble y presenta muchos valores faltantes.
* coords_modified: no nos da informacion relevante del apartamento
* piso: más del 98% de valores nulos, lo que impide imputación confiable.
* closets: también con más del 98% de valores nulos, sin utilidad para el modelo.

```python
columnas_eliminar = df_ini.columns[df_ini.isna().sum() <= 10]
columnas_eliminar
```

```python
filas_antes = df_ini.shape[0]
df_ini = df_ini.dropna(subset=columnas_eliminar).reset_index(drop=True)
filas_despues = df_ini.shape[0]

print("Filas eliminadas:", filas_antes - filas_despues)
```
